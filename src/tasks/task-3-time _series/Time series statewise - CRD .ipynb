```python
Trend = state_cases.groupby(['Date'])['Confirmed', 'Deceased','Recovered',].sum().reset_index()
```

```python
Trend.tail()
```

```python
fig = go.Figure(go.Bar(x= Trend.Date, y= Trend.Recovered, name='Recovered'))
fig.add_trace(go.Bar(x=Trend.Date, y= Trend.Deceased, name='Deceased'))
fig.add_trace(go.Bar(x=Trend.Date, y= Trend.Confirmed, name='Confirmed'))

fig.update_layout(barmode='stack',legend_orientation="h",legend=dict(x= 0.3, y=1.1),
                 paper_bgcolor='white',
                 plot_bgcolor = "white",)
fig.show()
```

```python
import plotly.express as px
```

```python
def horizontal_bar_chart(df, x, y, title, x_label, y_label, color):
    fig = px.bar(df, x=x, y=y, orientation='h', title=title, 
                 labels={x.name: x_label,
                         y.name: y_label}, color_discrete_sequence=[color])
    fig.update_layout(yaxis={'categoryorder': 'total ascending'})
    fig.show()
```

```python
top_15_death_states = india_cases.sort_values('Deceased',ascending = False)[1:16]
horizontal_bar_chart(top_15_death_states,top_15_death_states.Deceased, top_15_death_states.index,
                     'Top 15 States with most deaths','Number of deaths(In Thousands)','State Name','Red')
```

```python
top_10_confirmed_states = india_cases.sort_values('Confirmed', ascending=False)[1:11]
horizontal_bar_chart(top_10_confirmed_states,top_10_confirmed_states.Confirmed, top_10_confirmed_states.index,
            'Top 10 Indian States (Confirmed Cases)', 'Number of Confirmed cases (in Miliions)','States Name','blue')
```

```python
top_10_recovered_states = india_cases.sort_values('Recovered', ascending=False)[1:11]
horizontal_bar_chart(top_10_recovered_states, top_10_recovered_states.Recovered, top_10_recovered_states.index,
                    'Top 10 States (Recovered Cases)', 'Number of Recovered cases (in Millions)', 'States Name', 'Green')
```

```python
from fbprophet import Prophet
from fbprophet.plot import plot_plotly, add_changepoints_to_plot
from plotly.offline import iplot, init_notebook_mode
```

```python
model = Prophet()
```

```python
Confirmed = Trend.loc[:, ['Date', 'Confirmed']] 
Confirmed.tail()
```

```python
Recovered = Trend.loc[:, ['Date', 'Recovered']] 
Recovered.tail()
```

```python
Deceased = Trend.loc[:, ['Date', 'Deceased']] 
Deceased.tail()
```

```python
Confirmed.columns = ['ds', 'y']
model.fit(Confirmed)
```

```python
future = model.make_future_dataframe(periods=90) # helper function to extend the dataframe for specified days
future.tail()
```

```python
# yhat represents the prediction, while yhat_lower and yhat_upper represent the lower and upper bound
forecast_india_conf = model.predict(future)
forecast_india_conf
```

```python
fig = plot_plotly(model, forecast_india_conf) 

fig.update_layout(template='plotly_white')

iplot(fig) 
```
