### Import historical data

```python
import requests
import pandas as pd

response = requests.get("https://api.covid19india.org/v4/min/timeseries.min.json")
if response.status_code == 200:
    historical_day_numbers = response.json()
    DATE = []
    STATE = []
    TOTAL_CASES = []
    RECOVERED = []
    DECEASED = []
    TESTED = []
    VACCINATED_DOSE1 = []
    VACCINATED_DOSE2 = []
    
       
    for state in historical_day_numbers.keys():
        for date in historical_day_numbers[state]['dates'].keys():
            STATE.append(state)
            DATE.append(date)
            #Using try and except since in some dates cases calues are not available. So, except adds a zero where cases are not
            #available
            try:
                TOTAL_CASES.append(historical_day_numbers[state]['dates'][date]['delta']['confirmed'])
            except:
                TOTAL_CASES.append(0)
            try:
                RECOVERED.append(historical_day_numbers[state]['dates'][date]['delta']['recovered'])
            except:
                RECOVERED.append(0)
            try:
                DECEASED.append(historical_day_numbers[state]['dates'][date]['delta']['deceased'])
            except:
                DECEASED.append(0)
            try:
                TESTED.append(historical_day_numbers[state]['dates'][date]['delta']['tested'])
            except:
                TESTED.append(0)
            try:
                VACCINATED_DOSE1.append(historical_day_numbers[state]['dates'][date]['delta']['vaccinated1'])
            except:
                VACCINATED_DOSE1.append(0)
            try:
                VACCINATED_DOSE2.append(historical_day_numbers[state]['dates'][date]['delta']['vaccinated2'])
            except:
                VACCINATED_DOSE2.append(0)

            
    Covid19_historical_data = pd.DataFrame(
        {
            
            "STATE/UT": STATE,
            "DATE": DATE,
            "TOTAL_CASES": TOTAL_CASES,
            "RECOVERED": RECOVERED,
            "DECEASED": DECEASED,
            "TESTED": TESTED,
            "VACCINATED_DOSE1": VACCINATED_DOSE1,
            "VACCINATED_DOSE2": VACCINATED_DOSE2,
       }
    )
    #print(data.head())

else:
    print("Error while calling API: {}".format(response.status_code, response.reason))
```

```python
#Changing DATE column format
Covid19_historical_data['DATE']= pd.to_datetime(Covid19_historical_data['DATE'], format = '%Y-%m-%d')
```
